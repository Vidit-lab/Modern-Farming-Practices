```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('FarmSucessCategorized-1.xlsx')

df.head(40)
```

```python
df1 = pd.read_csv('Dataset05 - Sheet2.csv')
df1.head(40)
```

```python
# Checking count of crop_category
df1['crop_category'].value_counts()
```

```python
df1['crop_category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Distribution of Product Innovations by Crop Category")
```

Specialty crops dominate value-added innovation.

```python
df1['innovation_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Distribution of Product Innovations by Crop Category")
plt.show()
```

```python
df1['primary_crop'].value_counts()
```

```python
sns.boxplot(
    data=df1,
    x="crop_category",
    y="annual_revenue"
)
```

```python
# Grouping by innovation_type and calculating mean annual revenue upto 2 decimal places

mean_revenue_by_innovation = df1.groupby('innovation_type')['annual_revenue'].mean().reset_index().round(2)
print(mean_revenue_by_innovation) 
```

```python
# Grouping by crop_category and calculating mean annual_revenue and plotting it

mean_revenue_by_crop_category = df1.groupby('crop_category')['annual_revenue'].mean().reset_index().round(2)
sns.barplot(data=mean_revenue_by_crop_category, x='crop_category', y='annual_revenue')
plt.title('Mean Annual Revenue by Crop Category')
plt.xlabel('Crop Category')
plt.ylabel('Mean Annual Revenue')
plt.xticks(rotation=45)
plt.show()
 
```

```python
#Plotting bar plot for mean annual revenue by innovation type
sns.barplot(
    data=mean_revenue_by_innovation,
    x="innovation_type",
    y="annual_revenue"
)
plt.title("Mean Annual Revenue by Innovation Type")
plt.xlabel("Innovation Type")
plt.ylabel("Mean Annual Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# Countplot of states
plt.figure(figsize=(12, 6))
sns.countplot(data=df1, x='state')
plt.title('Count of Records by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
